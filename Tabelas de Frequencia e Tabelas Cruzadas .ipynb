```python
import pandas as pd
```

```python
from google.colab import drive

drive.mount('/content/drive')

netflix = pd.read_csv('/content/drive/MyDrive/Arquivos/netflix.csv')
```

```python
#TABELAS DE FREQUÊNCIA
#Gerar lista de Paísesflix
listapaises = list(netflix['production_countries'])
listapaisesunicos = list(set(netflix['production_countries']))
#Comando "set" cria um conjunto de elementos únicos
```

```python
#Fazer contagem em todos os Países
tabfrequencia = netflix['production_countries'].value_counts()
```

```python
#Transformar em DataFrame e nomear coluna
tabfrequencia = pd.DataFrame(tabfrequencia)
tabfrequencia.columns = ['Frequência']
tabfrequencia
```

```python
#Gerar Percentuais
tabfrequencia['%'] = netflix['production_countries'].value_counts(normalize=True)*100
tabfrequencia = tabfrequencia.round(2)
tabfrequencia
#"normalize=True" já adiciona o numero em percentual
```

```python
#TABULAÇÃO CRUZADA
pd.crosstab(netflix.production_countries,netflix.type)
```

```python
#Gerar Total
pd.crosstab(netflix.production_countries,netflix.type, margins = True)
#"margins = True" adiciona a coluna e a linha "Total"
```

```python
#Proporções
pd.crosstab(netflix.production_countries,netflix.type, margins = True, normalize = True)*100
```

```python
pd.crosstab(netflix.production_countries,netflix.type, margins = True, normalize = 'columns')*100
#"normalize = 'columns'" gera o percentual em relação a coluna
```

```python
#inserir um terceira Variável
```

```python
tabfreq = pd.crosstab(netflix.production_countries,[netflix.type,netflix.imdb_score.round(0)], margins = True, normalize = 'index')*100
tabfreq.round(2)
```
